```python
import numpy as np
import pandas as pd
from datetime import datetime

# European Files
df_cases = pd.read_csv("cases_deaths/May_daily_number_of.csv")
df_variants = pd.read_csv("variants/variants in the EU.csv")

# Unique country values (3stellige ISO)
cases_countries = sorted(df_cases["countryterritoryCode"].unique())
var_country_count = sorted(df_variants["country_code"].unique())

# Print all countries
print(cases_countries)
```

```python
# Load and show Kaggle dataset 
df_vacc = pd.read_csv("country_vaccinations_July21.csv")
df_vacc.head()
```

```python
# Filter from Kaggel vacc only the 30 European Countries and print unique values
df_europ_vacc = df_vacc[df_vacc["iso_code"].isin(cases_countries)]
df_europ_vacc["iso_code"].unique()
# 30 -> should be right 
```

```python
# reset index to range from 0 to ... 
df_europ_vacc.reset_index(drop=True, inplace=True)
df_europ_vacc
```

```python
# how many entries does each country have?
df_europ_vacc["iso_code"].value_counts()
# doesn't look to bad
```

```python
# check what the format of the dates is
vacc_dates = df_europ_vacc["date"][0]
print(type(vacc_dates))
print(vacc_dates)

# -> string -> need to convert to pandas datetime
```

```python
vacc_dates = pd.to_datetime(vacc_dates)
df_europ_vacc.index = pd.to_datetime(df_europ_vacc["date"])
df_europ_vacc.head()
```

```python
# after droping date column
df_europ_vacc.drop(columns=["date"], inplace=True)
df_europ_vacc
```

```python
# check if index is datetime
df_europ_vacc.index
# -> DateTimeIndex -> success
```

```python
# Check for null values
df_europ_vacc.isnull().sum()
# daily vaccinations only 30 null values -> good 
```

```python
# dealing with the time intervall
# start from 01.03.2021
condition1 = df_europ_vacc.index >= pd.Timestamp(2021, 7, 2)
df_europ_vacc = df_europ_vacc.loc[condition1]
df_europ_vacc.head()
```

```python
# check how many values each country now has -> should be the same for each
df_europ_vacc["iso_code"].value_counts()
# almost correct -> no real outliers anymore 
```

```python
# cut off timeintervall at the 06.05.2021 to get equal amounts of columns
condition2 = df_europ_vacc.index <= pd.Timestamp(2021, 7, 2)
df_europ_vacc = df_europ_vacc.loc[condition2]
df_europ_vacc.tail()
```

```python
# check value counts for each country again -> should be the same for each this time
df_europ_vacc.iso_code.value_counts()
# 67 for each -> success
```

```python
# next daily number of cases has to be set to the same time intervall
# then merge on columns iso_code (country) and date
# check documentation for merge (for additional arguments and formatting)

# check daily number of cases dataset -> dates are strings
type(df_cases.dateRep.iloc[0])
```

```python
# set date as datetime index and drop date column
# corrected it to the right format
dates = df_cases.dateRep
datetimes = pd.to_datetime(dates, format="%d/%m/%Y")
df_cases.dateRep = datetimes
df_cases
```

```python
# switch around datetime index -> start march end may
df_cases = df_cases.iloc[::-1]
df_cases
 
```

```python
# reset index of vaccination dataframe
df_europ_vacc = df_europ_vacc.reset_index (level='date')
df_europ_vacc
```

```python
# merge on date and country code
newframe2 = pd.merge(df_cases, df_europ_vacc, left_on=["dateRep","countryterritoryCode"], right_on=["date", "iso_code"], how="inner")
newframe2
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vaccinations_df = pd.read_csv('newframe_shortened2.csv')
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x=vaccinations_df.dateRep, y=vaccinations_df.daily_vaccinations)
plt.title('The Number of daily vaccinations dynamic')
plt.show()
```

```python
countries = vaccinations_df.groupby('countryterritoryCode')['daily_vaccinations'].max().sort_values(ascending= False)[:5].index

top_countries = pd.DataFrame(columns= vaccinations_df.columns)
for country in countries:
  top_countries = top_countries.append(vaccinations_df.loc[vaccinations_df['countryterritoryCode'] == country])
```

```python
# Explore the Vaccination procedure go on rapidly from which date.

plt.figure(figsize=(15,8))
sns.lineplot(top_countries['dateRep'], 
             top_countries['daily_vaccinations_per_million'], 
             hue= top_countries['countryterritoryCode'], ci= False)
plt.title('Vaccination procedure go on rapidly');
```

```python
# Which country has most number of fully vaccinated people?¶

fully_vaccinated = vaccinations_df.groupby("countryterritoryCode")["people_fully_vaccinated"].max().sort_values(ascending= False).head(25)
```

```python
fully_vaccinated.reset_index()
```

```python
plt.figure(figsize=(16,10))
ax = sns.barplot(x=fully_vaccinated, y=fully_vaccinated.index)
plt.xlabel("Fully Vaccinated")
plt.ylabel("Country");
plt.title('Which country has most number of fully vaccinated people?');

for patch in ax.patches:
    width = patch.get_width()
    height = patch.get_height()
    x = patch.get_x()
    y = patch.get_y()
    
    plt.text(width + x, height + y, '{:.1f} '.format(width))
```

```python
# Daily COVID-19 vaccine doses administered per million people.

daily_vaccinations_per_million = vaccinations_df.groupby("countryterritoryCode")["daily_vaccinations_per_million"].max().sort_values(ascending= False).head(15)
```

```python
daily_vaccinations_per_million.reset_index()
```

```python
plt.figure(figsize=(12,8))
ax = sns.barplot(x=daily_vaccinations_per_million, y=daily_vaccinations_per_million.index )
plt.xlabel("daily vaccinations per million")
plt.ylabel("Country")
plt.title("Daily COVID-19 vaccine doses administered per million people");

for patch in ax.patches:
    width = patch.get_width()
    height = patch.get_height()
    x = patch.get_x()
    y = patch.get_y()
    
    plt.text(width + x, height + y, '{:.1f} '.format(width))
```

```python
# How many people daily vaccinated in Italy?

ita_df = vaccinations_df[vaccinations_df['countryterritoryCode'] == 'ITA']
ita_df
```

```python
ita_df.daily_vaccinations.sum()
```

```python
plt.figure(figsize=(16,16))
sns.lineplot(x=ita_df.dateRep, y=ita_df.daily_vaccinations)
plt.xlabel("Date")
plt.ylabel("Daily_Vaccination")
plt.title('How many people daily vaccinated in Italy?');
```

```python
# What is the country that vaccinated completely most of the population

population_country=vaccinations_df.groupby('countryterritoryCode')['total_vaccinations_per_hundred'].max().sort_values(ascending=False).head(15)

population_country.reset_index()
```

```python
plt.figure(figsize= (15, 8))
ax = sns.barplot(x=population_country, y=population_country.index)
plt.title('Total Vaccinations / Population')
plt.xlabel('Total Vaccinations')
plt.ylabel('Country')

for patch in ax.patches:
    width = patch.get_width()
    height = patch.get_height()
    x = patch.get_x()
    y = patch.get_y()
    
    plt.text(width + x, height + y, '{:.1f} %'.format(width))
```
